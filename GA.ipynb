```python
#GA 遺伝的アルゴリズム実装
import random
import copy
gene_length=30#遺伝子長さ
individual_length=30
generation=40#世代数
mutate_rate=0.1#突然変異
selection_rate=0.2


#個体数
#populationは0 or 1でencoding
def get_population():
    population=[]
    for i in range(individual_length):
        population.append([random.randint(0,1) for j in range(gene_length)])
    return population

pop1=get_population()
pop1.sort()
pop1.reverse()
print(pop1)

#適合度を計算
def fitness(pop):
    return sum(pop)
#評価
def evaluate(pop):
    pop.sort()
    pop.reverse()
    return pop

#二点交差 子供を作る
def cross_over(parent1,parent2):
    r1=random.randint(0,gene_length-1)
    r2=random.randint(r1,gene_length-1)
    child=copy.deepcopy(parent1)
    child[r1:r2]=parent2[r1:r2]
    return child

def mutate(parent):
    r=random.randint(0,gene_length-1)
    child=copy.deepcopy(parent)
    #突然変異
    if child[r]==0:
        child[r]=1
    else:
        child[r]=0
    return child

#初期固体生成
pop=evaluate([(fitness(p),p) for p in get_population()])
#世代によってloop
for g in range(generation):
    eva=evaluate(pop)
    #elite selection
    elite=eva[:int(len(pop)*selection_rate)]
    print(elite)
    #突然変異 選択した個体の中から変異
    pop=elite
    while len(pop)<individual_length:
        if random.random()<mutate_rate:
            m=random.randint(0,len(elite)-1)
            child=mutate(elite[m][1])
        else:
            m1=random.randint(0,len(elite)-1)
            m2=random.randint(m1,len(elite)-1)
            child=cross_over(elite[m1][1],elite[m2][1])
        pop.append((fitness(child),child))
    
    eva=evaluate(pop)
    pop=eva
    
    print('Min:{}'.format(pop[-1][0]))
    print('max:{}'.format(pop[0][0]))
    print('================')
print('result:{}'.format(pop[0]))
```
