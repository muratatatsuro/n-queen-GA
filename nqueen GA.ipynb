```python
def show(a):
    n=len(a)
    for i in range(n):
        s=['-']*n
        s[a[i]]='Q'
        print(''.join(s))
    print
def _backtrack(a,n):
    if len(a)==n:
        show(a)
        return
    for s in range(n):#次の行のs列目においてみる
        #すでにｓ列目においてある
        if s in a:
            continue
        #斜めにおいてある
        if any([len(a)-i==abs(a[i]-s) for i in range(len(a))]):
            continue
        a.append(s)
        _backtrack(a,n)
        a.pop()
def back_track(n):
    _backtrack([],n)
    
back_track(10)
    
    
```
